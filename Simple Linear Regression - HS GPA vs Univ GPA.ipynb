```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading data
data = pd.read_csv('hs_univ.csv')
X = data['high_GPA'].values
Y = data['univ_GPA'].values
```

```python
#plotting data
plt.scatter(X, Y)
plt.show()
```

```python
#Simple Linear Regression using Ordinary Least Squares method
X_mean = np.mean(X)
Y_mean = np.mean(Y)
corr_X_Y = sum(X*Y)/np.sqrt(sum(X**2)*sum(Y**2))
X_std_dev = np.sqrt(sum((X-X_mean)**2))
Y_std_dev = np.sqrt(sum((Y-Y_mean)**2))
```

```python
#finding slope and intercept
slope = corr_X_Y * (Y_std_dev/X_std_dev)
y_int = Y_mean - slope * X_mean
print('Slope', slope)
print('Y_intercept', y_int)
```

```python
#plotting regression line
y = slope*X + y_int
plt.plot(X,y,'r')
plt.scatter(X,Y)
plt.xlabel('High School GPA')
plt.ylabel('University GPA')
plt.show()
```

```python
#Finding R_square value
y_dash = np.array([slope*X[i]+y_int for i in range(len(X))])
sum_squares = sum((Y-Y_mean)**2)
sum_residual_square = sum((Y-y_dash)**2)
R_square = 1 - (sum_residual_square/sum_squares)
print('R_square = ',R_square)
```

```python
#Finding Root Mean Squared Error - RMSE
RMSE = np.sqrt(sum((Y-y_dash)**2)/len(Y))
print('RMSE',RMSE)
```

```python
#Let us predict the University GPA if High School GPA is 3.6
x_test = 3.6
y_pred = slope * x_test + y_int
print('Predicted Univ GPA is ',y_pred)
```
